```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
```

# Question 1 - NN Basics
## I. Fitting Data
### Simulate DGP

```python
np.random.seed(123)
torch.manual_seed(123)
n = 400
x = np.random.uniform(0, 2*np.pi, n)
eps = np.random.normal(0, 0.2, n)
y = np.sin(x) + eps
```

```python
X = torch.tensor(x, dtype=torch.float32).view(-1, 1)
Y = torch.tensor(y, dtype=torch.float32).view(-1, 1)
```

```python
x_grid = np.linspace(0, 2*np.pi, 500)
X_grid = torch.tensor(x_grid, dtype=torch.float32).view(-1, 1)
true_grid = np.sin(x_grid)
```

#### Clase para red general

```python
class SimpleNN(nn.Module):
    def __init__(self, activation1, activation2=None, activation3=None):
        super().__init__()
        self.fc1 = nn.Linear(1, 50)
        self.fc2 = nn.Linear(50, 50)
        self.fc3 = nn.Linear(50, 50)
        self.out = nn.Linear(50, 1)

        # Activaciones
        self.act1 = activation1
        self.act2 = activation2 if activation2 else activation1
        self.act3 = activation3 if activation3 else activation1

    def forward(self, x):
        x = self.act1(self.fc1(x))
        x = self.act2(self.fc2(x))
        x = self.act3(self.fc3(x))
        return self.out(x)

def train_model(model, X, Y, epochs=2000, lr=0.01):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for _ in range(epochs):
        optimizer.zero_grad()
        y_pred = model(X)
        loss = criterion(y_pred, Y)
        loss.backward()
        optimizer.step()

    with torch.no_grad():
        mse = criterion(model(X), Y).item()

    return mse

def plot_fit(model, title):
    with torch.no_grad():
        y_grid_pred = model(X_grid).numpy().flatten()

    plt.figure(figsize=(8,4))
    plt.scatter(x, y, alpha=0.4, label="Datos simulados")
    plt.plot(x_grid, true_grid, "--", label="sin(x) verdadero")
    plt.plot(x_grid, y_grid_pred, label="Predicci√≥n NN")
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()
```

#### Entrenar modelos individuales

```python
results = {}

# Logistic (sigmoid)
model_log = SimpleNN(nn.Sigmoid())
mse_log = train_model(model_log, X, Y)
plot_fit(model_log, f"NN con activaci√≥n LOGISTIC - MSE={mse_log:.4f}")
results["logistic"] = mse_log

# tanh
model_tanh = SimpleNN(nn.Tanh())
mse_tanh = train_model(model_tanh, X, Y)
plot_fit(model_tanh, f"NN con activaci√≥n TANH - MSE={mse_tanh:.4f}")
results["tanh"] = mse_tanh

# ReLU
model_relu = SimpleNN(nn.ReLU())
mse_relu = train_model(model_relu, X, Y)
plot_fit(model_relu, f"NN con activaci√≥n ReLU - MSE={mse_relu:.4f}")
results["relu"] = mse_relu
```

#### Red con activaciones distintas

```python
model_mixed = SimpleNN(
    activation1 = nn.ReLU(),
    activation2 = nn.Tanh(),
    activation3 = nn.Sigmoid()
)

mse_mixed = train_model(model_mixed, X, Y)
plot_fit(model_mixed, f"NN con activaciones MIXTAS - MSE={mse_mixed:.4f}")
results["mixed"] = mse_mixed
```

#### Ajusta mejor

```python
print("MSE de cada modelo:")
for k, v in results.items():
    print(f"{k}: {v:.4f}")

best = min(results, key=results.get)
print(f"\n La red que mejor ajusta es: {best}")
```

# II. Learning-rate

#### ¬øQu√© es learning rate teoricamente?

##### El learning rate es un hiperpar√°metro que controla el tama√±o del paso que da el algoritmo de optimizaci√≥n (como Adam o SGD) en cada iteraci√≥n al actualizar los pesos de la red neuronal. Un learning rate muy peque√±o hace que el entrenamiento sea lento y puede atascarse, uno muy grande puede causar inestabilidad y que la red nunca converja. 

#### Simular

```python
np.random.seed(123)
torch.manual_seed(123)

n = 400
x = np.random.uniform(0, 2*np.pi, n)
eps = np.random.normal(0, 0.2, n)
y = np.sin(x) + eps

X = torch.tensor(x, dtype=torch.float32).view(-1, 1)
Y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

x_grid = np.linspace(0, 2*np.pi, 500)
X_grid = torch.tensor(x_grid, dtype=torch.float32).view(-1, 1)
true_grid = np.sin(x_grid)
```

#### Definimos clases y funciones

```python
class NN(nn.Module):
    def __init__(self, layers):
        super().__init__()
        self.layers = nn.ModuleList()
        for i in range(len(layers)-1):
            self.layers.append(nn.Linear(layers[i], layers[i+1]))
        self.activation = nn.ReLU()

    def forward(self, x):
        for layer in self.layers[:-1]:
            x = self.activation(layer(x))
        return self.layers[-1](x)
def train_model(model, lr, epochs=2000):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    for _ in range(epochs):
        optimizer.zero_grad()
        y_pred = model(X)
        loss = criterion(y_pred, Y)
        loss.backward()
        optimizer.step()

def get_pred(model):
    with torch.no_grad():
        return model(X_grid).numpy().flatten()
```

```python
lrs = [0.0001, 0.001, 0.01, 0.1]
```

#### Experimento 1

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y, alpha=0.3, label="Datos")

for lr in lrs:
    model = NN([1, 50, 1])
    train_model(model, lr)
    y_pred = get_pred(model)
    plt.plot(x_grid, y_pred, label=f"lr = {lr}")

plt.plot(x_grid, true_grid, "--", label="sin(x)")
plt.title("Learning rate comparison ‚Äî 1 hidden layer (ReLU)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### Experimento 2

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y, alpha=0.3, label="Datos")

for lr in lrs:
    model = NN([1, 50, 50, 1])
    train_model(model, lr)
    y_pred = get_pred(model)
    plt.plot(x_grid, y_pred, label=f"lr = {lr}")

plt.plot(x_grid, true_grid, "--", label="sin(x)")
plt.title("Learning rate comparison ‚Äî 2 hidden layers (ReLU)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### Experimento 3

```python
plt.figure(figsize=(10,5))
plt.scatter(x, y, alpha=0.3, label="Datos")

for lr in lrs:
    model = NN([1, 50, 50, 50, 1])
    train_model(model, lr)
    y_pred = get_pred(model)
    plt.plot(x_grid, y_pred, label=f"lr = {lr}")

plt.plot(x_grid, true_grid, "--", label="sin(x)")
plt.title("Learning rate comparison ‚Äî 3 hidden layers (ReLU)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### What can you say about the relationship between the learning rate and the number of hidden layers?

##### A medida que aumentamos el n√∫mero de capas ocultas, la red se vuelve m√°s profunda y, por lo tanto, m√°s dif√≠cil de optimizar. En modelos sencillos, como una red con una sola capa oculta, es com√∫n que learning rates relativamente altos, como 0.01 o incluso 0.1, funcionen bien y permitan que el entrenamiento converja r√°pido hacia una buena soluci√≥n. Sin embargo, cuando la red incorpora dos o tres capas ocultas, la situaci√≥n cambia: esos mismos learning rates altos suelen producir inestabilidad, haciendo que las predicciones oscilen, diverjan o se ajusten de forma muy irregular. Esto ocurre porque los gradientes deben propagarse a trav√©s de m√°s capas y cualquier actualizaci√≥n demasiado grande puede amplificar errores en lugar de corregirlos. En cambio, cuando utilizamos learning rates m√°s peque√±os, como 0.001 o 0.0001, el entrenamiento de redes profundas se vuelve mucho m√°s estable. Aunque el avance es m√°s lento, la red ajusta sus par√°metros de manera m√°s controlada y es capaz de aproximar mejor la forma suave de la funci√≥n objetivo. En conjunto, estos resultados muestran que, a medida que aumenta la profundidad de la red, es necesario reducir el learning rate para evitar inestabilidad y asegurar una convergencia adecuada.
